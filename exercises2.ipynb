# 5. Ice Hockey
Code below reads to DataFrame Finnish Hockey League (regular season) results from season 2010-2011 to  2019-2020.  

Fields:
* date
* game (home team - away team)
* result (home team goals - away team goals)
* outcome: JA if game has ended on overtime, VL if game has ended in shoot-out, otherwise empty

Winning team gets 3 points ans losing 0 points, except that if game has ended on overtime or shootout winner gets 2 and loser 1 points.


Write a function that takes a string team name (without extra spaces) as parameter and calculates and prints:
* team name
* total number of games for the team
* total number of wins for the team
* total number of losses for the team
* total number of points for the team
* number of goals for the team
* number of goals against the team

Example output:
```
Team: HIFK
Games: 599
Wins: 340
Losses: 259
Points: 1021
Goals for: 1713
Goals against: 1458
```

````python
df=pd.read_csv('http://gpspekka.kapsi.fi/dataanalytics/part2/liiga.txt', sep=';')

df[['eka','toka']] = df.game.str.split("-",expand=True,)
df['eka'] = df['eka'].str.strip()
df['toka'] = df['toka'].str.strip()
df['outcome'] = df['outcome'].str.strip()
df[['goals_for_t1','goals_for_t2']] = df.result.str.split("-",expand=True,)

print(df)
````

````python
def get_info(team):
    
    team1 = df['eka'] == team
    team2 = df['toka'] == team
    
    first_wins = df['goals_for_t1'] > df['goals_for_t2']
    second_wins = df['goals_for_t1'] < df['goals_for_t2']
    overtime_or_shootout = (df['outcome'] == "VL") | (df['outcome'] =="JA")
    
    games = df[team1 | team2]
    wins = df[( team1 & first_wins ) | ( team2 & second_wins )]
    losses = df[( team1 & second_wins ) | ( team2 & first_wins )]
  
    # overtime and shootous
    losses_os = df[(team1 & second_wins & overtime_or_shootout) | (team2 & first_wins & overtime_or_shootout)]
    wins_os = df[(team1 & first_wins & overtime_or_shootout)] + df[(team2 & second_wins & overtime_or_shootout)]
    points = 3 * (wins.shape[0] - wins_os.shape[0]) + 2 * wins_os.shape[0] + 1 * losses_os.shape[0]
    
    goals_for = df[team1]['goals_for_t1'].astype(int).sum() + df[team2]['goals_for_t2'].astype(int).sum()
    goals_against = df[team1]['goals_for_t2'].astype(int).sum() + df[team2]['goals_for_t1'].astype(int).sum()
    
    print("Team: " + team)
    print("Games: " + str(games.shape[0]))
    print("Wins: " + str(wins.shape[0]))
    print("Losses: " + str(losses.shape[0]))
    print("Points: " + str(points))
    print("Goals for: "+  str(goals_for))
    print("Goals against: " + str(goals_against))
    
    
get_info("HIFK")
````
